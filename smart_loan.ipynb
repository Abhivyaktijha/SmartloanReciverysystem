This notebook is prepared so that **plots show up on GitHub**. GitHub does **not** execute notebooks; it only displays **saved outputs**.

### How to make sure your plots are visible on GitHub
1. **Run your notebook locally or in Colab** before pushing, so outputs (images) are saved inside the `.ipynb`.
2. For **Matplotlib/Seaborn**, just keep `%matplotlib inline` (default in Jupyter) and run the cells. The output PNGs will be saved in the notebook.
3. For **Plotly**, set the renderer to **static PNG** so GitHub can show it:
   ```python
   import plotly.io as pio
   pio.renderers.default = 'png'  # requires kaleido
   ```
   If you see an error, install Kaleido in the environment **before** running the plotting cells:
   ```python
   %pip install -U kaleido
   ```
4. Commit/push the **executed** notebook (with outputs).

````
# --- Plot Setup Helpers ---
%matplotlib inline
try:
    import plotly.io as pio
    # If Kaleido is available, PNG rendering will work; otherwise Plotly will error on show() with 'png'.
    pio.renderers.default = 'png'
except Exception as e:
    print('Plotly PNG renderer not set. If you plan to use Plotly on GitHub, run: %pip install -U kaleido')
````

````
# --- Example Matplotlib Plot ---
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x) * np.cos(3*x/2)
plt.figure(figsize=(6,3.5), dpi=120)
plt.plot(x, y, lw=2)
plt.title('Example Matplotlib Plot â€” Saved as PNG in Output')
plt.xlabel('x')
plt.ylabel('sin(x)Â·cos(1.5x)')
plt.grid(True, alpha=0.4)
plt.show()
````

### Example Plotly Usage (Static PNG Renderer)
Below is a template. If Kaleido is installed, the figure will render as a static PNG in the output. Run the two cells (setup and this one) before pushing to GitHub.

If you get an error like `ValueError: Image export using the 'kaleido' engine requires the kaleido package`, install it with `%pip install -U kaleido` and re-run.

````
# --- Example Plotly Plot (Static PNG) ---
import plotly.graph_objects as go
import numpy as np
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines', name='sin(x)'))
fig.update_layout(title='Example Plotly Figure â€” PNG Renderer for GitHub', xaxis_title='x', yaxis_title='sin(x)')
fig.show()  # will display as PNG if pio.renderers.default = 'png' and kaleido is installed
````
